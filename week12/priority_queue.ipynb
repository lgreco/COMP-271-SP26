# Priority queues


## Complete binary tree

A complete binary tree is a binary tree in which every level except possibly the deepest, is completely filled. At the last level, all nodes must be as far left as possible. [[NIST definition.](https://xlinux.nist.gov/dads/HTML/completeBinaryTree.html)]

> A binary tree is a two-dimensional data structure comprising nodes. Each node may have 0, 1, or 2 links to other nodes immediately below it. The links cannot form cycles.


Both binary trees below are complete with depth $L=3$. The tree to the left is the smallest possible complete binary tree with 3 levels. It has $2^{L-1}$ nodes. The tree to the right is the fullest complete binary tree possible with 3 levels. It has $2^L-1$ nodes.

![](https://raw.githubusercontent.com/lgreco/images/refs/heads/main/pqueue/pqueue_cbt.png)

In general, the number $n$ of nodes in a complete binary tree of depth $L$ is

$$
2^{L-1} \leq n \leq 2^L-1
$$

Conversely, the depth $L$ of a complete binary tree with $n$ nodes is

$$
L = \lfloor \log_2 n \rfloor +1
$$

A tree with depth $L$ has at most $2^L-1$ nodes. This predictable upper limit allows us to use a humble array of length $2^L-1$ as an equivalent storage for the data contained in the tree.

![](https://raw.githubusercontent.com/lgreco/images/refs/heads/main/pqueue/pqueue_tree_array.png)

The parent-children links can be preserved/recovered with simple arithmetic. For a node placed in position $i$ in the array of length $n$, its children and parent are given by:

$$
\begin{align*}
{\color{red}\text{left\_child\_of}}(i) & = 2i+1 ,&\, & 0 \leq i < \lfloor n/2 \rfloor & \\
{\color{green}\text{right\_child\_of}}(i) & = 2i+2 ,&\, & 0 \leq i < \lfloor n/2 \rfloor &  \\
{\color{blue}\text{parent\_of}}(i) & = \left\lfloor\frac{i-1}{2}\right\rfloor,&\, & 0 < i \leq n &
\end{align*}
$$

It's worth noting that the operation $\lfloor\frac{i-1}{2}\rfloor$ is implemented as integer division. For example in Python
```python
def parent(child:int) -> int:
    return (child - 1) // 2 
```

## Swap

![](https://raw.githubusercontent.com/lgreco/images/refs/heads/main/pqueue/pqueue_swap.png)

Using array representation for a complete binary tree, swapping a parent with one of its children is trivial. For the swap indicared by the purple arrow above, we can simply write

```python
def swap(parent: int, child: int) -> None
    array[parent], array[child] = array[child], array[parent]
```

The syntax above is quite Pythonic, so it may be better to define a standard swap:

````python
def swap(i: int, j: int, array) -> None:
    """Swaps two elements in given array using a temp. variable"""
    # Guard statement
    if (0 <= i < len(array)) and (0 <= j < len(array)):
        # Safe to execute
        temp = array[i]
        array[i] = array[j]
        array[j] = temp
````

And so the swap shown by the purple arrow above, can be written as:

```python
swap(parent, left_child_of(parent), array)
```


## Swap by property

In the swaps above, we are targeting a specific child. What if we do not know which child node we wish to swap, and instead we want to swap a parent node with the smaller or the larger of its two children?

![](https://raw.githubusercontent.com/lgreco/images/refs/heads/main/pqueue/pqueue_swap_select.png)





````python
def swap_smallest_child_of(parent: int, array: list) -> None:
    # Guard statement
    if 0 <= parent < len(array) // 2:
        # Safe to execute. Assume smallest is left
        smallest: int = left_child_of(parent)
        # smallest+1 is the right child
        if array[smallest + 1] < array[smallest]:
            # The right child is smaller
            smallest = smallest + 1
        swap(parent, smallest)
````

## Why is this important?

Consider a complete binary tree structure where every "family" of nodes observes the following rule:

*The value of the parent node is greater than or equal to the values of both its children.*

In the complete binary tree below there are three such families, outlined in red, blue, and green. 

![](https://raw.githubusercontent.com/lgreco/images/refs/heads/main/pqueue/pqueue_basic_cbt.png)

In each of the three families above, the parent is greater than or equal to the values of both its children. In mathematical terms,

$$
\begin{align*}
{\text{parent}}    & \ge \max\left\{ \text{right}, \text{left}\right\} \\
{\color{red}100}\  & {\color{red}  {\ge \max \left\{ 45, 50 \right\} }} \\
{\color{green}45}\ & {\color{green}{\ge \max \left\{ 35, 45 \right\} }} \\
{\color{blue}50}\  & {\color{blue} {\ge \max \left\{ 48, 45 \right\} }}
\end{align*}
$$

A complete binary tree structured in this fashion, has a special trait: it's largest value is always at the top. 

> Oddly enough, the top of a tree in computer science is called the **root** of the tree.



The requirement that ${\text{parent}} \ge \max\left\{ \text{right}, \text{left}\right\}$ is called the **maximum heap property.** If it is maintained by every node in a complete binary tree, except the childless nodes, the tree is said to be a **maximum heap.**

> Childless nodes in a binary tree are called **leaf nodes.**

Conversely, the requirement that ${\text{parent}} \le \min\left\{ \text{right}, \text{left}\right\}$ is called the **minimum heap property.** If it is maintained by every node in a complete binary tree, except the childless nodes, the tree is said to be a **minimum heap.**

In a maximum heap, the highest valued node is the root node. In a minimum heap, the root node has the lowest value. 

When a heap is implemented using an array, as shown earlier, the root node is the first element of the array at position `[0]`. The last non empty element of the array is the bottom-righmost node of the tree. If the tree has $n$ nodes, the last non empty element is at position `[n-1]`.
