```python
from __future__ import annotations  # Authorized import for advanced type hints

# 2345678901234567890123456789012345678901234567890123456789012345678901234567890


################################################
# SOLUTIONS CODE - DO NOT SHARE ON PUBLIC REPO #
################################################


class Node:
    """A simple linkable object. The node comprises two fields: a value field,
    here typed a string, and a pointer field to the next node. The default is
    to instantiate a node with just a value, and no other node to point to it,
    for example:

    +------+
    | node |-next--> None
    +------+

    Nodes are connected to each other later, for example,

    chi = Node("Chicago")        spi = Node("Springfield")
    +------+                     +------+
    | node |-next--> None        | node |-next--> None
    +------+                     +------+

    and then chi.set_next(spi) will result to the arrangement below which is,
    essentially, a linked list.

    chi -----------------------> spi

    +------+                     +------+
    | node |-next--------------> | node |-next--> None
    +------+                     +------+

    """

    def __init__(self, value: str) -> None:
        """Object constructor. Requires only a value to instantiate the object.
        The next object may be assigned later"""
        self.__value: str = value
        self.__next: Node | None = None

    def __str__(self) -> str:
        """Simple string representation of the object."""
        return f"{self.__value}"

    def has_next(self) -> bool:
        """Predicate accessor; tells if object points to
        another object"""
        return self.__next is not None

    def get_next(self) -> Node | None:
        """Standard accessor for next object"""
        return self.__next

    def set_next(self, next: Node) -> None:
        """Mutates object by assigning its next pointer to another object"""
        self.__next = next

    def get_value(self) -> str:
        """Accessor for the object's value"""
        return self.__value


################################################
# SOLUTIONS CODE - DO NOT SHARE ON PUBLIC REPO #
################################################


class LinkedList:
    """A simple linked list of Node objects. Nodes in this list are
    connected one after the other, as shown below

      head
    +------+         +------+         +------+        +------+
    | node |-next--> | node |-next--> | node |--> ... | node |-next-->  None
    +------+         +------+         +------+        +------+

    Every node, in the linked list, is connected to another node. Except for
    the last node that points to None.

    PROBLEM 1: SOLUTION requires adding a tail pointer to the linked list,
    that always points to the last node in the list. This way, we can add a
    new node to the end of the list in constant time, without having to traverse
    the entire list to find the last node. The tail pointer is updated every
    time a new node is added to the end of the list.

    POBLEM 2: SOLUTION requires adding a count variable to the linked list,
    that keeps track of the number of nodes in the list. This way, we can
    return the number of nodes in the list in constant time, without having to
    traverse the entire list to count the nodes. The count variable is updated
    every time a new node is added to the list, and every time a node is
    removed from the list.
    """

    def __init__(self) -> None:
        """Instantiate an empty linked list"""
        # At its most elemental form, a linked list is just a pointer to
        # the first node in the list, which is called the head of the list.
        # If the list is empty, the head is None.
        self.__head: Node | None = None
        # PROBLEM 1: SOLUTION requires adding a tail pointer to the linked
        # list, which is a pointer to the last node in the list.
        # The tail pointer is initialized to None, because the list is empty
        # at the time of instantiation.
        self.__tail: Node | None = None
        # PROBLEM 2: SOLUTION requires adding a count variable to the linked
        # list, which is an integer that keeps track of the number of nodes
        # in the list. The count variable is initialized to 0, because the
        # list is empty at the time of instantiation.
        self.__count: int = 0

    # Constants for string representation of the linked list
    __EMPTY_LIST_STR: str = "Empty List"
    __RIGHT_ARROW: str = " → "

    def __str__(self) -> str:
        """String representation."""
        # Assume an empty list
        string: str = self.__EMPTY_LIST_STR
        if self.__head is not None:
            # If the list is not empty, we start with the head node and
            # we keep adding the value of the next node, until we reach
            # the end of the list.
            string = self.__head.get_value()
            # Start with the first node and keep adding the next node
            # until we reach the end of the list.
            current: Node = self.__head.get_next()
            while current is not None:
                # Add the next node to the string and move to the next
                # node. The loop ends when we reach the end of the list
                # and current is None.
                string += self.__RIGHT_ARROW + current.get_value()
                current = current.get_next()
        return string

    def add(self, value: str) -> None:
        """Adds a new node to the linked list. First we create a new node
        with the given value. And next we find the end of the linked list
        and we append the new node to it."""
        # Create the new node to be added
        new_node: Node = Node(value)
        # If the list is empty, the new node becomes its head and tail
        if self.__head is None:
            self.__head = new_node
        else:
            # If the list is not empty, we append the new node to the tail
            self.__tail.set_next(new_node)
        # PROBLEM 1: SOLUTION requires updating the tail pointer to point to
        # the new node, which is now the last node in the list.
        self.__tail = new_node
        # Update the count of nodes in the list
        self.__count += 1

    def count(self) -> int:
        """Counts the number of nodes in the object and returns it"""
        return self.__count

    def insert(self, new_value: str, after_value: str) -> bool:
        """Inserts a new node with the given value after the first node that
        has the after_value. Returns True if the insertion was successful, and
        False if there is no node with the after_value in the list."""
        # Start at the head of the list and look for the node with the
        # after_value
        current: Node | None = self.__head
        # Initialize a flag to indicate if the insertion was successful
        success = False
        # Loop through the list until we find the node with the
        # after_value or we reach the end of the list
        while current is not None and not success:
            if current.get_value() == after_value:
                # We found the node with the after_value, so we create
                # a new node with the new_value and we insert it after
                # the current node
                new_node: Node = Node(new_value)
                # We set the next pointer of the new node to point to the next
                # node of the current node, and we set the next pointer of
                # the current node to point to the new node. This way, we
                # insert the new node in the list after the current node.
                new_node.set_next(current.get_next())
                # We set the next pointer of the current node to point to
                # the new node
                current.set_next(new_node)
                # If we are inserting after the tail, we need to update
                # the tail pointer
                if current == self.__tail:
                    self.__tail = new_node
                # Update the count of nodes in the list
                self.__count += 1
                # Set the success flag to True and break the loop
                success = True
            current = current.get_next()
        return success
```

## Unit testing for part A

```python
import unittest


class TestLinkedListInsert(unittest.TestCase):
    # --- helpers -------------------------------------------------------------

    def make_list(self, *values: str):
        ll = LinkedList()
        for v in values:
            ll.add(v)
        return ll

    # --- tests ---------------------------------------------------------------

    def test_insert_into_empty_list_returns_false_and_no_change(self):
        ll = LinkedList()
        self.assertEqual("Empty List", str(ll))
        self.assertEqual(0, ll.count())

        result = ll.insert("X", "A")
        self.assertFalse(result)
        self.assertEqual("Empty List", str(ll))
        self.assertEqual(0, ll.count())

    def test_insert_missing_after_value_returns_false_and_no_change(self):
        ll = self.make_list("A", "B", "C")
        before_str = str(ll)
        before_count = ll.count()

        result = ll.insert("X", "NOPE")
        self.assertFalse(result)
        self.assertEqual(before_str, str(ll))
        self.assertEqual(before_count, ll.count())

    def test_insert_after_head(self):
        ll = self.make_list("A", "B", "C")
        result = ll.insert("A2", "A")

        self.assertTrue(result)
        self.assertEqual("A → A2 → B → C", str(ll))
        self.assertEqual(4, ll.count())

    def test_insert_in_middle(self):
        ll = self.make_list("A", "B", "C")
        result = ll.insert("B2", "B")

        self.assertTrue(result)
        self.assertEqual("A → B → B2 → C", str(ll))
        self.assertEqual(4, ll.count())

    def test_insert_after_tail_updates_end_of_list(self):
        ll = self.make_list("A", "B", "C")
        result = ll.insert("C2", "C")

        self.assertTrue(result)
        self.assertEqual("A → B → C → C2", str(ll))
        self.assertEqual(4, ll.count())

        # A follow-up insert-after-tail should now treat C2 as the tail
        result2 = ll.insert("C3", "C2")
        self.assertTrue(result2)
        self.assertEqual("A → B → C → C2 → C3", str(ll))
        self.assertEqual(5, ll.count())

    def test_insert_after_first_match_only_when_duplicates_exist(self):
        ll = self.make_list("A", "B", "C", "B")  # duplicate B at the end
        result = ll.insert("X", "B")

        self.assertTrue(result)
        # Must insert after the FIRST "B", not the last one
        self.assertEqual("A → B → X → C → B", str(ll))
        self.assertEqual(5, ll.count())

    def test_failed_insert_does_not_mutate_list(self):
        ll = self.make_list("A", "B", "C")
        snapshot_str = str(ll)
        snapshot_count = ll.count()

        self.assertFalse(ll.insert("Y", "ZZZ"))
        self.assertEqual(snapshot_str, str(ll))
        self.assertEqual(snapshot_count, ll.count())


# If you test in a .PY file, uncomment the following
# two lines to run the tests.

# if __name__ == "__main__":
#    unittest.main()

# If you test in a Jupyter notebook, uncomment the following
# line to run the tests.
unittest.main(argv=[''], exit=False)
```

```python
from __future__ import annotations  # Authorized import for advanced type hints
from abc import ABC, abstractmethod  # Authorized import for derived classes

# 2345678901234567890123456789012345678901234567890123456789012345678901234567890


################################################
# SOLUTIONS CODE - DO NOT SHARE ON PUBLIC REPO #
################################################


class Performance(ABC):
    """
    A general live performance event.

    This class captures the shared structure and behavior of all
    live performances: concerts, lectures, theater productions,
    magic shows, etc.

    The purpose of this class is to define:

        • Common data that every performance has.
        • Common behavior shared by all performances.
        • A contract (via abstract methods) that subclasses must fulfill.

    Subclasses are responsible for defining how revenue is calculated
    and how the performance is described.
    """

    def __init__(
        self, title: str, duration_minutes: int, base_ticket_price: float
    ) -> None:
        """
        Initialize a new performance.

        Parameters:
            title               The name of the event.
            duration_minutes    How long the event lasts.
            base_ticket_price   The standard ticket price before
                                any subclass-specific adjustments.

        Note:
            We use protected attributes (_name style) instead of
            private (__name) because subclasses will need direct
            access to these values.
        """
        self._title: str = title
        self._duration_minutes: int = duration_minutes
        self._base_ticket_price: float = base_ticket_price

        # Number of audience members currently admitted.
        # Starts at zero and increases via admit_audience().
        self._audience_count: int = 0

    # ---------------------------------------------------------
    # Concrete (Fully Implemented) Methods
    # These are inherited as-is by subclasses.
    # ---------------------------------------------------------

    def __str__(self) -> str:
        """
        General string representation.

        We call describe() here so that when a Performance
        object is printed, the subclass version of describe()
        is used automatically (polymorphism in action).
        """
        return self.describe()

    def admit_audience(self, number: int) -> None:
        """
        Adds audience members to the performance.

        Only positive numbers are accepted.
        """
        if number > 0:
            self._audience_count += number

    def get_title(self) -> str:
        """Returns the performance title."""
        return self._title

    def get_duration(self) -> int:
        """Returns the duration in minutes."""
        return self._duration_minutes

    def get_audience_count(self) -> int:
        """Returns the number of admitted audience members."""
        return self._audience_count

    def get_base_ticket_price(self) -> float:
        """
        Returns the base ticket price.

        Subclasses may use this value as the starting point
        for their own pricing logic.
        """
        return self._base_ticket_price

    # ---------------------------------------------------------
    # Abstract Methods (Must Be Implemented by Subclasses)
    # ---------------------------------------------------------

    @abstractmethod
    def calculate_revenue(self) -> float:
        """
        Compute total revenue for the performance.

        Subclasses decide how ticket price is adjusted
        (VIP upgrades, student discounts, special pricing, etc.).

        The result should reflect:
            audience_count × adjusted_ticket_price
        """
        ...

    @abstractmethod
    def describe(self) -> str:
        """
        Return a human-readable description of the performance.

        Each subclass should include details specific
        to its type of event.
        """
        ...
```

```python
class Concert(Performance):
    """
    A live music concert class derived from the Performance class.
    """

    # VIP upgrade increases ticket price by 40%
    __VIP_PRICE_MULTIPLIER: float = 1.4

    def __init__(
        self,
        title: str,
        duration_minutes: int,
        base_ticket_price: float,
        artist_name: str,
        genre: str,
        has_vip: bool,
    ) -> None:
        """Initialize a new concert."""
        # Call the superclass constructor to initialize common fields
        super().__init__(title, duration_minutes, base_ticket_price)
        # Initialize subclass-specific fields
        self._artist_name: str = artist_name
        self._genre: str = genre
        self._has_vip: bool = has_vip

    def calculate_revenue(self) -> float:
        # Start with the base ticket price and adjust it based
        # on VIP status
        adjusted_price = self.get_base_ticket_price()
        if self._has_vip:
            # Increase by 40%
            adjusted_price *= self.__VIP_PRICE_MULTIPLIER
        return self.get_audience_count() * adjusted_price

    # Constant for string representation of the concert
    __CONCERT_STR_TEMPLATE: str = "Concert '{}' by {} ({} genre), lasting {} minutes."

    def describe(self) -> str:
        # Provide a textualdescription that includes the
        # artist name and genre
        return self.__CONCERT_STR_TEMPLATE.format(
            self.get_title(),
            self._artist_name,
            self._genre,
            self.get_duration(),
        )

    def has_vip(self) -> bool:
        """Returns True if the concert has VIP tickets, False otherwise."""
        return self._has_vip


class Lecture(Performance):
    """
    An educational lecture derived from the Performance class.
    """

    def __init__(
        self,
        title: str,
        duration_minutes: int,
        base_ticket_price: float,
        speaker_name: str,
        has_student_discount: bool,
    ) -> None:
        """Initialize a new lecture."""
        # Call the superclass constructor to initialize common fields
        super().__init__(title, duration_minutes, base_ticket_price)
        # Initialize subclass-specific fields
        self._speaker_name: str = speaker_name
        self._has_student_discount: bool = has_student_discount

    # Student discount decreases ticket price by 50%
    __STUDENT_DISCOUNT_MULTIPLIER: float = 0.50

    def calculate_revenue(self) -> float:
        """Start with the base ticket price and adjust it based
        on student discount status"""
        adjusted_price = self.get_base_ticket_price()
        if self._has_student_discount:
            # Decrease by 25%
            adjusted_price *= self.__STUDENT_DISCOUNT_MULTIPLIER
        return self.get_audience_count() * adjusted_price

    # Constant for string representation of the lecture
    __LECTURE_STR_TEMPLATE: str = "Lecture '{}' by {}, lasting {} minutes."

    def describe(self) -> str:
        """
        Returns a string description of the lecture.
        """
        return self.__LECTURE_STR_TEMPLATE.format(
            self.get_title(), self._speaker_name, self.get_duration()
        )

    def has_student_discount(self) -> bool:
        """Returns True if the lecture has a student discount, False otherwise."""
        return self._has_student_discount
```
