# Week 05: Abstract Data Types


```python
class Character:
    """A class to represent a character in a show."""

    def __init__(self, first_name: str, last_name: str, role: str):
        """Initialize a Character object with the specified first name,
        last name, and role."""
        self.__first_name = first_name
        self.__last_name = last_name
        self.__role = role

    def __str__(self) -> str:
        """Returns a string representation of the Character object. Specifically,
        the string consists of the first letter of the first name, last name,
        and role concatenated together.
        """
        return (
            ("" if self.__first_name == "" else self.__first_name[0])
            + ("" if self.__last_name == "" else self.__last_name[0])
            + ("" if self.__role == "" else self.__role[0])
        )

    # 2345678901234567890123456789012345678901234567890123456789012345678901234567890

    def get_first_name(self) -> str:
        """Returns the first name of the character."""
        return self.__first_name

    def get_last_name(self) -> str:
        """Returns the last name of the character."""
        return self.__last_name

    def get_role(self) -> str:
        """Returns the role of the character."""
        return self.__role

    def __eq__(self, other) -> bool:
        # A guard statement to make sure that other is of type Character
        if not isinstance(other, Character):
            raise ValueError("not same objects")
        return (
            self.__first_name == other.get_first_name()
            and self.__last_name == other.get_last_name()
        )
    
    def __lt__(self, other) -> bool:
        """Less-than operation between characters, alphabetically
        by first name."""
        return self.__first_name < other.get_first_name()
```

```python
rs1 = Character("River", "Song", "Time Traveler")
rs2 = Character("Leo", "Irakliotis", "Professor")
```

```python
rs1 > rs2
```
